```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def decay(x,*popt):
    a, b, c = popt
    
    return a*np.exp(-x/b)+c
```

```python
data = np.loadtxt('data.data')
events = []
times = []
for i in range(0, len(data)):
    if data[i][0] < 40000:
        events.append(data[i][0])
        times.append((data[i][1]-data[0][1])/(3600))
```

```python
nb=20
x = np.linspace(0, 20000, 200)
eventshist = plt.hist(events, bins=np.linspace(0,20000,nb), label = 'Event Histogram')
posx = np.array(eventshist[1][:-1]) + (eventshist[1][1]/2)
posy = np.array(eventshist[0])
yerror = np.sqrt(np.array(eventshist[0]))
yerror[13]=1.0
plt.errorbar(posx,posy,yerr=yerror, linestyle='None', label='Error Bars', capsize = 7.5)
plt.legend(loc='best')
plt.xlabel(r'Decay time (ns)')
plt.ylabel('Count/Bin')
```
